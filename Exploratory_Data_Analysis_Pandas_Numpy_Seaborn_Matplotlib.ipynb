```
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

file_name = "https://raw.githubusercontent.com/maryamsyed/Data-Science-and-Deep-Learning-for-Business/master/winequalityN.csv"
df = pd.read_csv(file_name)
df.head()
```

```
df.describe()
```

```
df.info()
```

```
df.hist(column='alcohol', bins = 15, grid = False, figsize=(10,6),color = '#86bf91')
```

```
sns.distplot(df['alcohol'],bins = 25, kde = True)
```

```
_ = sns.distplot(df.alcohol)
_ = plt.xlabel('Alcohol Percentage')
_ = plt.ylabel('Count')
_ = plt.title('Alcohol Content')
```

```
sns.lmplot(x='alcohol',y='fixed acidity',data=df)
```

```
sns.lmplot(x = 'alcohol', y='chlorides',data=df, fit_reg = False)
```

```
sns.lmplot(x='alcohol',y='chlorides', data = df, fit_reg = False, hue = 'type')
```

```
print(df["alcohol"].quantile(0.1))
print(df["alcohol"].quantile(0.5))
print(df["alcohol"].quantile(0.9))
print(df["alcohol"].quantile(0.99))
```

```
df["alcohol"].quantile([0.5,0.9])
```

```
sns.boxplot(data=df['alcohol'])
```

```
plt.figure(figsize=(12,8))

summary = df.drop(['free sulfur dioxide','total sulfur dioxide','quality','residual sugar','fixed acidity','alcohol'],axis=1)

sns.boxplot(data=summary,palette='Set3')
```

```
sns.violinplot(data=df['alcohol'])
```

```
sns.kdeplot(df.alcohol,df.density)
```

```
sns.jointplot(x='alcohol',y='density',data=df)
```

```
x = np.sort(df['alcohol'])
y = np.arange(1,len(x)+1) / len(x)
_ = plt.plot(x,y,marker=".",linestyle='none')
plt.margins(0.05)
_=plt.xlabel('Percent of Alcohol in Wine')
_=plt.ylabel('ECDF')
plt.grid(True)
plt.show()
```
